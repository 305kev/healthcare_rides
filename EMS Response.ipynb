```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/EMSResponses.csv")
df.head()
```

```python
#Remove all of the temporal and non-unique data that we don't care about.
df.drop(["census_state_code", "census_county_code", "census_block_name", "zone"], axis = 1, inplace = True)
```

```python
df.drop(["incident_date", "time_phone_pickup", "time_unit_arrived","time_incident_closed"], axis = 1, inplace = True)
```

```python
df["disposition"][df["disposition"] == "<Missing>"] = "Other"
```

```python
df["census_geoID"] = df["census_geoID"].map(lambda x: int(x/10000))
```

```python
df.head()
df_EMS = df[df["disposition"] == "ATCEMS Transport"]
df_EMS = df_EMS.groupby("census_geoID").size().reset_index(name = "count")
```

```python
df_population = pd.read_csv("data/PopulationData.csv")
df_population = df_population[["GEO.id2", "HC01_VC03"]]
df_population = df_population.iloc[1:].astype(int)
```

```python
df_EMS.merge(df_population, left_on = "census_geoID", right_on = "GEO.id2", how = 'inner')
```

```python
df.to_csv("ReducedEMSResponse.csv")
```

```python
df2 = df.groupby("disposition").size().reset_index(name="total_incidents")
df2["average_per_month"] = (df2["total_incidents"] / 58).round()
df2["average_per_year"] = df2["average_per_month"] * 12
df2
```
