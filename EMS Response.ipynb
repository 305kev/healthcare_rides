```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/EMSResponses.csv")
df.head()

```

```python
#Remove all of the temporal and non-unique data that we don't care about.
df.drop(["census_state_code", "census_county_code", "census_block_name", "zone"], axis = 1, inplace = True)
```

```python
df.drop(["incident_date", "time_phone_pickup", "time_unit_arrived","time_incident_closed"], axis = 1, inplace = True)
```

```python
df["disposition"][df["disposition"] == "<Missing>"] = "Other"
```

```python
df.info()
```

```python
df2 = df.groupby("disposition").size().reset_index(name="total_incidents")
df2["average_per_month"] = (df2["total_incidents"] / 58).round()
df2["average_per_year"] = df2["average_per_month"] * 12
df2
```
